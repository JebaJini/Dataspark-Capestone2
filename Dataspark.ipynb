```python
# Creating AGE Group
query= """SELECT 
    CustomerKey,
    Name,
    Birthday,
    TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) AS Age,
    CASE
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 0 AND 18 THEN '0-18'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 19 AND 35 THEN '19-35'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 36 AND 50 THEN '36-50'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 51 AND 65 THEN '51-65'
        ELSE '65+'
    END AS Age_Group
FROM 
    dataspark;"""
result= pd.read_sql(query,engine)
print(result)
```

```python
# Products bought by Age Group
query="""SELECT 
    Category,
    CASE
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 0 AND 18 THEN '0-18'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 19 AND 35 THEN '19-35'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 36 AND 50 THEN '36-50'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 51 AND 65 THEN '51-65'
        ELSE '65+'
    END AS Age_Group,
    COUNT(*) AS Total_Products_Bought
FROM 
    dataspark
GROUP BY 
    Category, Age_Group
ORDER BY 
    Age_Group, Category;"""
result= pd.read_sql(query,engine)
print(result)
```

```python
#Inactive customers by gender
query="""SELECT 
    Gender,
    COUNT(*) AS Total_Customers
FROM (
    SELECT 
        CustomerKey,
        Gender,
        COUNT(CustomerKey) AS Total_Orders
    FROM 
        dataspark
    GROUP BY 
        CustomerKey, Gender
    HAVING 
        COUNT(CustomerKey) = 1
) AS subquery
GROUP BY 
    Gender;"""
result= pd.read_sql(query, engine)
print(result)
```

```python
# Age Group bought max products
query="""SELECT 
    CASE
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 0 AND 18 THEN '0-18'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 19 AND 35 THEN '19-35'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 36 AND 50 THEN '36-50'
        WHEN TIMESTAMPDIFF(YEAR, Birthday, CURDATE()) BETWEEN 51 AND 65 THEN '51-65'
        ELSE '65+'
    END AS Age_Group,
    SUM(Quantity) AS Total_Products_Bought
FROM 
    dataspark
GROUP BY 
    Age_Group
ORDER BY 
    Total_Products_Bought DESC
LIMIT 10;"""
result= pd.read_sql(query,engine)
print(result)
```

```python
#On Time Delivery Percentage
query= """SELECT 
    COUNT(*) AS Total_Orders,
    SUM(CASE WHEN `Delivery Date` <= `Order Date` THEN 1 ELSE 0 END) AS On_Time_Deliveries,
    (SUM(CASE WHEN `Delivery Date` <= `Order Date` THEN 1 ELSE 0 END) / COUNT(*)) * 100 AS On_Time_Percentage
FROM 
    dataspark;"""
result=pd.read_sql(query,engine)
print(result)
```

```python
#Monthly Sales Revenue
query="""SELECT 
    YEAR(`Order Date`) AS Year,
    MONTH(`Order Date`) AS Month,
    SUM(`Quantity` * `Unit Price USD`) AS Total_Revenue
FROM 
    dataspark
GROUP BY 
    YEAR(`Order Date`), MONTH(`Order Date`)
ORDER BY 
    Year, Month;"""
result=pd.read_sql(query,engine)
print(result)
```
